```python
import os
import random

# Define paths to the subdirectories
dataset_path = "./lung-and-colon-cancer-histopathological-images/lung_colon_image_set"
subdirectories = ["lung_image_sets/lung_n", "lung_image_sets/lung_aca", "lung_image_sets/lung_scc",
                  "colon_image_sets/colon_n", "colon_image_sets/colon_aca"]

# Seed for reproducibility (optional)
random.seed(42)

# Select and output 20% of images from each subdirectory
for subdir in subdirectories:
    subdir_path = os.path.join(dataset_path, subdir)
    print(f"\nSelecting 20% of images from: {subdir}")

    # Get the list of image files
    image_files = [f for f in os.listdir(subdir_path) if f.lower().endswith(('.png', '.jpg', '.jpeg'))]

    # Calculate 20% of the total images
    sample_size = max(1, int(len(image_files) * 0.2))  # Ensure at least 1 image is selected
    sampled_images = random.sample(image_files, sample_size)

    # Output the selected image file paths
    for image_file in sampled_images:
        image_path = os.path.join(subdir_path, image_file)
       # print(f"Selected Image: {image_path}")
```

```python
import os
import tensorflow as tf
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras import layers, models
from sklearn.model_selection import train_test_split

# Define the dataset path and classes
dataset_path = "./lung-and-colon-cancer-histopathological-images/lung_colon_image_set"
class_directories = [
    "lung_image_sets/lung_n",
    "lung_image_sets/lung_aca",
    "lung_image_sets/lung_scc",
    "colon_image_sets/colon_n",
    "colon_image_sets/colon_aca"
]

# Prepare data: Collect file paths and labels
image_paths = []
labels = []
for idx, class_dir in enumerate(class_directories):
    class_path = os.path.join(dataset_path, class_dir)
    for image_file in os.listdir(class_path):
        if image_file.lower().endswith(('.png', '.jpg', '.jpeg')):
            image_paths.append(os.path.join(class_path, image_file))
            labels.append(idx)

# Split data into training and testing sets
train_paths, test_paths, train_labels, test_labels = train_test_split(
    image_paths, labels, test_size=0.2, random_state=42, stratify=labels)

# Define a function to load and preprocess images
def preprocess_image(image_path, label, target_size=(128, 128)):
    image = tf.io.read_file(image_path)
    image = tf.image.decode_jpeg(image, channels=3)
    image = tf.image.resize(image, target_size) / 255.0  # Normalize to [0, 1]
    return image, label

# Create TensorFlow datasets
train_dataset = tf.data.Dataset.from_tensor_slices((train_paths, train_labels))
train_dataset = train_dataset.map(preprocess_image).shuffle(1000).batch(32)

test_dataset = tf.data.Dataset.from_tensor_slices((test_paths, test_labels))
test_dataset = test_dataset.map(preprocess_image).batch(32)

# Build a simple CNN model
model = models.Sequential([
    layers.Conv2D(32, (3, 3), activation='relu', input_shape=(128, 128, 3)),
    layers.MaxPooling2D((2, 2)),
    layers.Conv2D(64, (3, 3), activation='relu'),
    layers.MaxPooling2D((2, 2)),
    layers.Conv2D(128, (3, 3), activation='relu'),
    layers.MaxPooling2D((2, 2)),
    layers.Flatten(),
    layers.Dense(128, activation='relu'),
    layers.Dense(len(class_directories), activation='softmax')  # Output layer
])

# Compile the model
model.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])

# Train the model
print("\nTraining the model...")
history = model.fit(train_dataset, epochs=10, validation_data=test_dataset)

# Evaluate the model
print("\nEvaluating the model...")
test_loss, test_accuracy = model.evaluate(test_dataset)
print(f"Test Accuracy: {test_accuracy * 100:.2f}%")
```

```python
import matplotlib.pyplot as plt
history_dict = history.history
# Plot accuracy
plt.figure(figsize=(5, 5))
plt.plot(history_dict['accuracy'], label='Training Accuracy')
plt.plot(history_dict['val_accuracy'], label='Validation Accuracy')
plt.title('Training and Validation Accuracy')
plt.xlabel('Epochs')
plt.ylabel('Accuracy')
plt.legend()
plt.show()
```

```python
# Plot loss
plt.figure(figsize=(5, 5))
plt.plot(history_dict['loss'], label='Training Loss')
plt.plot(history_dict['val_loss'], label='Validation Loss')
plt.title('Training and Validation Loss')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.legend()
plt.show()
```
